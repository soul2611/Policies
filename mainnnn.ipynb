```python
def date_lastmodified(new_soup):
    try:

        div_tags = new_soup.find_all("div", attrs={'id': 'texttospeak'})

        last_p_tag = div_tags[-1].find_all("p")[-1]

        if last_p_tag:
            extracted_content = last_p_tag.get_text(separator=' ', strip=True)
            return extracted_content
        else:
            return "No <p> tags found."
    except:
        return "Error occurred while extracting the modified date."
```

```python
# Function to extract  Title
def get_title(new_soup):
    try:
        # Outer Tag Object
        h3_tag = new_soup.find("h3", class_="card-title title")
        
        # Inner NavigatableString Object
        title_value = h3_tag.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string



# # Function to extract Last Modified Date
# def date_lastmodified(new_soup):
#     try:

#         div_tags = new_soup.find_all("div", attrs={'id': 'texttospeak'})

#         last_p_tag = div_tags[-1].find_all("p")[-1]

#         if last_p_tag:
#             extracted_content = last_p_tag.get_text(separator=' ', strip=True)
#             return extracted_content
#         else:
#             return "No <p> tags found."
#     except:
#         return "Error occurred while extracting the modified date."

    #     try:

#         div_tags = new_soup.find_all("div", attrs={'id': 'texttospeak'})

#         last_p_tag = div_tags[-1].find_all("p")[-1]

#         if last_p_tag:
#             extracted_content = last_p_tag.get_text(separator=' ', strip=True)
#             return extracted_content
#         else:
#             return "No <p> tags found."
#     except:
#         return "Error occurred while extracting the modified date."


# def get_lastmodified(new_soup):
#     try:
#         # Find all div tags with id 'MiddleColumn_internal'
#         div_tags = new_soup.find_all("div", attrs={'id': 'MiddleColumn_internal'})
        
#         # Get the last p tag within the specified divs
#         last_p_tag = div_tags[-1].find_all("p")[-0]

#         if last_p_tag:
#             # Extract the text content from the last <p> tag
#             extracted_content = last_p_tag.get_text(separator=' ', strip=True)
#             return extracted_content
#         else:
#             return "No <p> tags found."
#     except:
#         return "Error occurred while extracting the modified date."

# # Example usage:
# # modified_date = get_modified_date(new_soup)
# # print(modified_date)

# Function to extract Last Modified Date
def get_lastmodified(new_soup):
    try:
        # Find all div tags with id 'MiddleColumn_internal'
        div_tags = new_soup.find_all("div", attrs={'id': 'MiddleColumn_internal'})
        
        # Get the last p tag within the specified divs
        last_p_tag = div_tags[-1].find_all("p")[0]

        if last_p_tag:
            # Extract the text content from the last <p> tag
            extracted_content = last_p_tag.get_text(separator=' ', strip=True)
            return extracted_content
        else:
            return "No <p> tags found."
    except:
        return "Error occurred while extracting the modified date."

# Example usage:
# modified_date = get_lastmodified(new_soup)
# print(modified_date)




# Function to extract Scheme
def get_source(new_soup):
    try:
        # Find all div tags with id 'MiddleColumn_internal'
        div_tags = new_soup.find_all("div", attrs={'id': 'MiddleColumn_internal'})
        
        # Get the last p tag within the specified divs
        last_p_tag = div_tags[-1].find_all("p")[-1]

        if last_p_tag:
            # Extract the text content from the last <p> tag
            extracted_content = last_p_tag.get_text(separator=' ', strip=True)
            return extracted_content
        else:
            return "No <p> tags found."
    except:
        return "Error occurred while extracting the source of the scheme."

# Example usage:
# scheme_source = get_scheme(new_soup)
# print(scheme_source)




```

```python
import numpy as np
```

```python

if __name__ == '__main__':

    #HEADERS FOR REQUEST
    HEADERS = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36', 'Accept-Langugage':'en-US, en;q=0.5'})


    # The webpage URL
    URL = "https://vikaspedia.in/agriculture/policies-and-schemes/crops-related"

    # HTTP Request
    webpage = requests.get(URL, headers=HEADERS)

    # Soup Object containing all data
    soup = BeautifulSoup(webpage.content, "html.parser")

    # Fetch links as List of Tag Objects
    links = soup.find_all("a", attrs={'class': 'folderfile_name'})

    # Store the links
    links_list = []

    # Loop for extracting links from Tag Objects
    for link in links:
            links_list.append(link.get('href'))

    d = {"title":[], "details":[], "source":[], "date":[]}
    
    # Loop for extracting agriculture details from each link 
    for link in links_list:
        new_webpage = requests.get("https://vikaspedia.in" + link, headers=HEADERS)

        new_soup = BeautifulSoup(new_webpage.content, "html.parser")

        # Function calls to display all necessary product information
        d['title'].append(get_title(new_soup))
        d['details'].append(get_lastmodified(new_soup))
        d['source'].append(get_source(new_soup))
        d['date'].append(date_lastmodified(new_soup))


    
    amazon_df = pd.DataFrame.from_dict(d)
    amazon_df['title'].replace('', np.nan, inplace=True)
    amazon_df = amazon_df.dropna(subset=['title'])
    amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```

```python
import csv
import json

# Function to convert CSV to JSON
def csv_to_json(csv_file_path, json_file_path):
    data = []
    with open(csv_file_path, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            data.append(row)

    with open(json_file_path, 'w') as json_file:
        json.dump(data, json_file, indent=4)

# Example usage:
csv_file_path = 'amazon_data.csv'  # Path to your CSV file
json_file_path = 'amazon_data.json'  # Path where you want to save the JSON file
csv_to_json(csv_file_path, json_file_path)
```
