```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
URL = "https://vikaspedia.in/agriculture/policies-and-schemes/crops-related"
```

```python
#HEADERS FOR REQUEST
HEADERS = ({'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36', 'Accept-Langugage':'en-US, en;q=0.5'})
```

```python
#HTTP REQUEST
webpage = requests.get(URL, headers=HEADERS)
```

```python
webpage
```

```python
type(webpage.content)
```

```python
#SOUP OBJECT CONTAINING ALL DATA
soup = BeautifulSoup(webpage.content, "html.parser")
```

```python
soup
```

```python
#FETCH LINK AS LISTS OF TAG OBJECT
links = soup.find_all("a", attrs={'class': 'folderfile_name'})
```

```python
links
```

```python
link = links[0].get('href')
```

```python
product_list = "https://vikaspedia.in" + link
```

```python
product_list
```

```python
new_webpage = requests.get(product_list, headers=HEADERS)
```

```python
new_webpage
```

```python
#SOUP OBJECT CONTAINING ALL DATA
new_soup = BeautifulSoup(new_webpage.content, "html.parser")
```

```python
new_soup
```

```python
# VIMP
# EXTRACTING  ORIGINAL TITLE NAME
h3_tag = new_soup.find("h3", class_="card-title title")

if h3_tag:
    # Extract the text content from the h3 tag
    title_data = h3_tag.text.strip()
    print( title_data)
else:
    print("No <h3> tag found with class 'card-title title'.")
```

```python
#VIMP
# MODIFIED DATE OF SCHEME
# Extract the  last modified date of the scheme using index -3
new_soup.find_all("div", attrs={'id': 'MiddleColumn_internal'})
last_p_tag = new_soup.find_all("p")[-3]
if last_p_tag:
    # Extract the text content from the last <p> tag
    extracted_content = last_p_tag.get_text(separator=' ', strip=True)
    print( extracted_content)
else:
    print("No <p> tags found.")
```

```python
#VIMP
# SOURCE OF SHCEME 
#  Extract the source of the scheme using index -4
new_soup.find_all("div", attrs={'id': 'MiddleColumn_internal'})
last_p_tag = new_soup.find_all("p")[-4]
if last_p_tag:
    # Extract the text content from the last <p> tag
    extracted_content = last_p_tag.get_text(separator=' ', strip=True)
    print( extracted_content)
else:
    print("No <p> tags found.")
```
